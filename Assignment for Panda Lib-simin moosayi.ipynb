```python
import pandas as pd
import numpy as np
```

```python
#Q1:create a dataframe which contains information of a company's products:
```

```python
data=pd.DataFrame({'prod_name':['g1','g2','c1','c2','c3','f1','f2'],
                   'price':[45,25,75,35,89,20,55],
                   'monthly_demand':[1200,2500,200,1850,120,3500,545],
                   'brand':['z','x','x','y','z','y','z']})
data
```

```python
#a: Add a column to calculate monthly revenue of each product:
data['monthly_revenu']=data['price']*data['monthly_demand']
data
```

```python
# b: Calculate total monthly revenue generated by brand z:
sum(data.loc[data['brand']=='z','monthly_revenu'])
```

```python
#c: Create a column called revenue share which includes monthly revenue share of each product
data['revenue_share']=round(data['monthly_revenu']/sum(data['monthly_revenu'])*100,2)
data
```

```python
#d: Which brand does have the largest revenue share?
```

```python
unique=data['brand'].unique()
max=0
brand_max=None
for brand in unique:
    total=data.loc[data['brand']==brand,'revenue_share'].sum() #calculate sum of revenue share for each brand

if total>max:
    max=total
    brand_max=brand

brand_max
```
